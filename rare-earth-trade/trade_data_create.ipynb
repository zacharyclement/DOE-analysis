```python
#on concat, ignore index=True allows index to represent when new spreadsheet starts. False aligns index
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    clean_file = codecs.open(file_, 'r', encoding='utf-8', errors='ignore')
    df = pd.read_csv(clean_file, skiprows=1 ,names = ['Country', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', 'Exporter', 'Form'])
    df.columns
    list_.append(df)  
frame = pd.concat(list_, ignore_index=False)
```
